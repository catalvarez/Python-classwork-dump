```python
# imports
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['font.size'] = 11
import numpy as np
import scipy.constants as spc
```

```python
omega = np.linspace(1.0e9,np.pi*2.0e9,100000)
L = 1.0
R = np.array([0.5,0.9,0.99])
R = R.reshape(len(R),1)
T = 1.0-R
I_circ_inc = T/(1+R**2-2.0*R*np.cos(omega/(spc.c)*2.0*L))

fig = plt.figure(figsize=[14, 10])
gs = matplotlib.gridspec.GridSpec(2, 3, width_ratios=[20, 20, 20], wspace=0.4, hspace=0.4)
plt.subplot(gs[0])
plt.ylim([0,100])
plt.title('R = '+str(R[0]))
plt.plot(omega,I_circ_inc[0,:])
plt.subplot(gs[1])
plt.ylim([0,100])
plt.title('R = '+str(R[1]))
plt.plot(omega,I_circ_inc[1,:])
plt.subplot(gs[2])
plt.ylim([0,100])
plt.title('R = '+str(R[2]))
plt.plot(omega,I_circ_inc[2,:])
```

```python
I_trans_inc = T**2/(1+R**2-2.0*R*np.cos(omega/spc.c*2.0*L))

fig = plt.figure(figsize=[14, 10])
gs = matplotlib.gridspec.GridSpec(2, 3, width_ratios=[20, 20, 20], wspace=0.3, hspace=0.4)
plt.subplot(gs[0])
plt.ylim([0,2])
plt.title('R = '+str(R[0]))
plt.plot(omega,I_trans_inc[0,:])
plt.subplot(gs[1])
plt.ylim([0,2])
plt.title('R = '+str(R[1]))
plt.plot(omega,I_trans_inc[1,:])
plt.subplot(gs[2])
plt.ylim([0,2])
plt.title('R = '+str(R[2]))
plt.plot(omega,I_trans_inc[2,:])
```
