```python
# imports
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['font.size'] = 11
import numpy as np
import scipy.constants as spc
```

```python
Wp = np.arange(0.01,15,0.001)
g32 = 100.0
g21 = 1.0
g3 = 100.0
N = 100

ss_pop_diff = (g32*Wp-g21*g3-g21*Wp)/(g21*g3+2*g21*Wp+g32*Wp)*N

plt.plot(Wp,ss_pop_diff)
plt.xlabel('Pumping rate')
plt.ylabel('steady state population difference')
```

```python
kappa = 1.0e-2 #cm/W
g = 0.00503 #cm-1
I_sat = 1
T = 0.01
I_pump = np.arange(0,2,0.01)

I_out = (2.0*T)*I_sat/2.0*(kappa*I_pump/g-1.0)

plt.plot(I_pump,I_out)
plt.ylim([0,0.04])
plt.xlabel('I_pump (W/cm^2)')
plt.ylabel('I_out (W/cm^2)')
print 'slope efficiency', (2.0*T)*I_sat/2.0*kappa/g
print 'threshold pumping intensity', 0.5
```
