```python
# imports
import matplotlib
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [9, 6]
matplotlib.rcParams['font.size'] = 11
import numpy as np
```

```python
N = np.arange(0, 120, 1)
N_m = np.array([0,1.0,2.0,5.0,10,100])
N_m = N_m.reshape(len(N_m),1)

p_N = np.exp(-N_m)*N_m**N/scipy.misc.factorial(N)

n_max = np.argmax(p_N,1)

fig = plt.figure(figsize=[12, 10])
gs = matplotlib.gridspec.GridSpec(2, 3, width_ratios=[20, 20, 20], wspace=0.3, hspace=0.4)


plt.subplot(gs[0])
plt.ylim([0,0.5])
plt.grid()
plt.title('N_bar = '+str(N_m[0]) +', max = '+str(N[n_max[0]]))
plt.plot(N,p_N[0,:])
plt.subplot(gs[1])
plt.ylim([0,0.5])
plt.grid()
plt.title('N_bar = '+str(N_m[1]) +', max = '+str(N[n_max[1]]))
plt.plot(N,p_N[1,:])
plt.subplot(gs[2])
plt.ylim([0,0.5])
plt.grid()
plt.title('N_bar = '+str(N_m[2]) +', max = '+str(N[n_max[2]]))
plt.plot(N,p_N[2,:])
plt.subplot(gs[3])
plt.ylim([0,0.5])
plt.grid()
plt.title('N_bar = '+str(N_m[3]) +', max = '+str(N[n_max[3]]))
plt.plot(N,p_N[3,:])
plt.subplot(gs[4])
plt.ylim([0,0.5])
plt.grid()
plt.title('N_bar = '+str(N_m[4]) +', max = '+str(N[n_max[4]]))
plt.plot(N,p_N[4,:])
plt.subplot(gs[5])
plt.ylim([0,0.5])
plt.grid()
plt.title('N_bar = '+str(N_m[5]) +', max = '+str(N[n_max[5]]))
plt.plot(N,p_N[5,:])
```

As the average number of photons increases, the probability distribution spreads out. This makes sense, because there is more possibility of individual trial photon count variation in each trial from say, 100 photons +/- 10%, than there is in 1 photon +/- 10%.
